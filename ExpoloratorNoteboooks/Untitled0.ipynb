```python
import dendropy
```

```python
tree = dendropy.Tree.get_from_path('./Trees/garli_opt/2bnoterm', 'newick')
```

```python
pv_mrca = tree.mrca(taxon_labels=["Achalinus meiguensis", "Trimeresurus puniceus"])
```

```python
ser_ml=dendropy.Tree()
ser_ml.seed_node = pv_mrca.parent_node
ser_ml.write_to_path("ser_ml.tre","newick")
```

```python
ser_ml.as_ascii()
```

```python
ser_ml.print_plot()
```
